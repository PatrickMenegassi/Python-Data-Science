```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Criação de uma variavel com dados em uma distribuição mormal com a função rvs (100 elementos)
dados = norm.rvs(size = 1000)
dados
```

```python
# Histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
# Geração de grafico para verificar se a distribuição é normal
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
# Teste de Shapiro
stats.shapiro(dados)
```

```python
# Dados não normais
dados2 = skewnorm.rvs(4, size=1000)
```

```python
# Histograma
plt.hist(dados2, bins = 20)
plt.title('Dados2')
```

```python
# Teste de Shapiro
stats.shapiro(dados)
```
