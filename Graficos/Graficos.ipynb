```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
```

```python
base = pd.read_csv("trees.csv")
base.shape
```

```python
base.head()
```

```python
#Criação do histograma, considerando somente o segundo atributo da base de dados e com duas divisões (bins)
# A variavel 'h' as faixas de valores de Height
h = np.histogram(base.iloc[:,1], bins=6)
h
```

```python
#Visualizando o histograma com 6 divisoes
plt.hist(base.iloc[:,1], bins=6)
plt.title('Arvores')
plt.ylabel('Frequencia')
plt.xlabel('Altura')
```

```python
#Histograma com 10 divisões (bins) e somente para o primeiro atributo da base de dados
plt.hist(base.iloc[:,1], bins=6)
```

```python
#Histograma com a linha de distribuição de frequencia, com 6 divisões (bins)
#kde = linha de densidade
sns.histplot(base.iloc[:,1], kde=False, bins=6, color='blue').set(title='Arvores')
```

```python
#Densidade, método kdeplot para densidade
sns.kdeplot(base.iloc[:,1], color='blue').set(title="Arvores")
```

```python
#Densidade e histograma
sns.histplot(base.iloc[:,1], kde=True, bins=6, color='blue').set(title='Arvores')
```

```python
# Gráfico de dispersão, considerando o volume e a dispersão
plt.scatter(x=base.Girth, y=base.Volume, color='blue', facecolors = 'none', marker = '*')
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
#Grafico de linha considerando o volume e o atributo 'girth' - sem marker
plt.plot(base.Girth, base.Volume)
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
#Grafico de dispersão com 'afastamento' dos dados (jitter)
#fit_reg linha de tendencia
sns.regplot(x=base.Girth, y=base.Volume, data = base, x_jitter = 0.3, fit_reg = False)
```

```python
#Grafico de dispersão com legendas
base = pd.read_csv('co2.csv')
base.head()
```

```python
# Criando duas variaveis para cada atributo (x=conc e y=uptake)
x = base.conc
y = base.uptake
```

```python
#Retorna os valores unicos do atributo "treatment"
unicos = list(set(base.Treatment))
unicos
```

```python
#Percorre cada tipo de tratamento (chilled e nonchilled) e cria um grafico de dispersao
for i in range(len(unicos)):
    indice = base.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label = unicos[i])
plt.legend(loc = 'lower right')
```

```python
#Boxplot
#Geração do boxplot
#patch_artist = True preenche, showfliers outlires
base = pd.read_csv('trees.csv')
plt.boxplot(base.Volume, vert = False, showfliers = False, notch = True, patch_artist = True)
plt.title('Arvores')
plt.xlabel('Volume')
```

```python
#Dados por linha
plt.boxplot(base)
plt.title('Arvores')
plt.xlabel('Dados')
```

```python
#Geração de 3 boxplot, cada um mostrando informações diferentes
plt.boxplot(base.Volume, vert = False)
plt.boxplot(base.Height, vert = False)
plt.boxplot(base.Girth, vert = False)
plt.title('Arvores')
plt.xlabel('Dados')
```

```python
#Grafico de barras e de setores (pizza)
base = pd.read_csv('insect.csv')
base.shape
base.head()
```

```python
#Agrupamemto dos dados baseado no atributo 'spray', contando e somando os registros
agrupado = base.groupby(['spray'])['count'].sum()
agrupado
```

```python
#Grafico de barras
agrupado.plot.bar(color = 'gray')
```

```python
#Corres
agrupado.plot.bar(color = ['blue', 'yellow', 'red', 'green', 'pink', 'orange'])
```

```python
#Grafico de pizza
agrupado.plot.pie()
```

```python
#Com legenda
agrupado.plot.pie(legend=True)
```

```python
#Grafico em 3d
base = pd.read_csv('orchard.csv')
base
```

```python
#Criacao do grafico 3D, indicando o atributo projection = '3d' e passando tres atributos (decrease, rowpos e colpos)
figura = plt.figure()
eixo = figura.add_subplot(1,1,1,projection='3d')
eixo.scatter(base.decrease, base.rowpos, base.colpos)
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
```
